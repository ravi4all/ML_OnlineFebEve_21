```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
```

```python
df = pd.read_csv('ipl-data/IPL Matches 2008-2020.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
print(pd.value_counts(df['winner']))
plt.hist(pd.value_counts(df['winner']))
# plt.show()
```

```python
df_1 = pd.value_counts(df['winner'])
```

```python
df_1
```

```python
plt.figure(figsize=(12,7))
# plt.bar(df_1.index, df_1)
plt.barh(df_1.index, df_1)
```

```python
plt.figure(figsize=(12,7))
plt.bar(df_1.index, df_1)
plt.title("Total wins by team")
plt.xticks(rotation='vertical')
plt.xlabel("Teams")
plt.ylabel("Count")
plt.show()
```

```python
df_2 = pd.value_counts(df['city']).head()
```

```python
df_2
```

```python
plt.figure(figsize=(7,7))
plt.pie(df_2, labels=df_2.index, autopct='%1.2f%%',
        colors=['cyan','gold','gray','red','green'],
       explode=(0,0,0,0.2,0), counterclock=False, startangle=90)
plt.legend(loc='upper right')
plt.title("City Plot")
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.countplot(x=df['winner'], data=df)
plt.title("Total wins by team")
plt.xticks(rotation='vertical')
plt.xlabel("Teams")
plt.ylabel("Count")
plt.show()
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
# pd.to_datetime(matches['date']).dt.month
df['season'] = pd.to_datetime(df['date']).dt.year
```

```python
plt.figure(figsize=(12,7))
sns.countplot(x=df['season'], data=df, hue=df['toss_decision'])
plt.title("Total wins by team")
plt.xticks(rotation='vertical')
plt.xlabel("Teams")
plt.ylabel("Count")
plt.show()
```
