```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('headbrain.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
X = df['Head Size(cm^3)'].values
y = df['Brain Weight(grams)'].values
```

```python
mean_x = np.mean(X)
mean_y = np.mean(y)
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(10,6))
plt.scatter(X, y)
plt.xlabel('Head Size')
plt.ylabel('Brain Weight')
plt.title('Head Size vs Brain Weight')
plt.show()
```

```python
numer = 0
denom = 0
n = len(X)

for i in range(n):
    numer += (X[i] - mean_x) * (y[i] - mean_y)
    denom += (X[i] - mean_x) ** 2

slope = numer/denom

inter = mean_y - slope * mean_x
```

```python
print("Slope :",slope)
print("Intercept :",inter)
```

```python
predictions = []
for i in range(n):
    pred = slope * X[i] + inter
    predictions.append(int(pred))
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(10,6))
plt.scatter(X, y)
plt.plot(X, predictions, 'blue')
plt.xlabel('Head Size')
plt.ylabel('Brain Weight')
plt.title('Head Size vs Brain Weight')
plt.show()
```

```python
error = 0
for i in range(n):
    error += (predictions[i] - y[i]) ** 2
    
mse = error/n
print("MSE :",mse)
```

```python
mae = np.sqrt(mse)
print("MAE :",mae)
```
